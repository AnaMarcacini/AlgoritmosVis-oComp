```python
def DRAW_BOXES(img,boxes,confidences,class_ids):
  for box,confidence,class_id in zip(boxes,confidences,class_ids):
    x1,y1,x2,y2 = map(int,box)
    # cv2.rectangle(img,(x1,y1),(x+w,y+h),(0,255,0),2)
    cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
    # labeçl = f'{result[0].names[class_id]} {confidence:.2f}'
    # label = f'{result[0].names[class_id]}'
    label = f'{model.names[int(class_id)]}'
    cv2.putText(img,label,(x1,y1),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
```
